### Advanced Operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant id, 2021, 2022, 2023, 2024]
saled_data = np.array([
    [1, 150000, 180000, 220000, 240000],
    [2, 120000, 170000, 120000, 110000],
    [3, 140000, 140000, 320000, 230000],
    [4, 180000, 210000, 200000, 150000],
    [5, 160000, 185000, 290000, 250000],
])

print("=== Zomato sales analysis ===")
print("\nSales data shape: ", saled_data.shape)

# basically we are writing the value of the rows and column in differnt manner 

# here we are simply doing the slicing in which  we pass how many rows we want to show
print("Sample data for 3 restaurants: ", saled_data[:3]) # [0:3]


# for getting data of only starting 3 restaurant

# for getting the whole data but from 1 column not from the 0 column
# in rows we are passing : -- means all rows 
# for cols we are passing 1: -- means starting from 1 and goes upto end column
print("Data starting from 1 column: ", saled_data[:, 1:])
```

```python
## total sales per year 

# sum method is used for summing whole data 
sum = np.sum(saled_data)
print(sum)

# [:, 1:] -- this is used for excluding the first col which is id

# but we want only the sum of sales .. we do not want to include the id 
sum = np.sum(saled_data[:, 1:])
print(sum)

# now we want to sum all the restaurant saled in each year 
# this will sum col wise and the first col will not be summed 
sum = np.sum(saled_data[:, 1:], axis=0)
print(sum)

# if we want sum row wise .. like total sale of each restaurant 
# this will give sum of each hotel sale in all years excluding the id
sum = np.sum(saled_data[: ,1:], axis=1)
print(sum)


# gettin the min sale of each restaurant 
# in this we want the min sale done by each restaurant from all years .. here row wise data reading is needed
min = np.min(saled_data[: , 1:], axis=1)
print(min)


# if we want max sale in a year .. means in this we will consider every restaurant each time .. and will take the maximum of all of them in that year 
# here col wise data reading is needed
max = np.max(saled_data[:, 1:], axis=0)
print(max)

# average sale for each restaurant -- for this we have to move row wise 
mean = np.mean(saled_data[:, 1:], axis=1)
print(mean)

# cumulative sum or can say the adjacant sum 
cumsum = np.cumsum(saled_data[:, 1:], axis=1)
print(cumsum)
```

```python
# now as we have imported the matplotlib .. then we could make the graph by the help of it
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales across all restaurant ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

### Vector Operations

```python
#1D
arr1 = np.array([1,2,3,4])
arr2 = np.array([5,6,7,8])

#2D
Arr1 = np.array([[1,2,3,4], [5,6,7,8]]) # (2,4)
Arr2 = np.array([[2,4,6,8], [7,4,3,1]]) # (2,4)

# addition -- in this each index element is added with the same index element in the another vector 
print("1D vector addition: ", arr1 + arr2)
print("2D vector addition: ", Arr1 + Arr2)

# multiplication -- in this each element is multiplied with the same index element in another vector
print("1D vector multiplication: ", arr1 * arr2)
print("2D vector multiplication: ", Arr1 * Arr2)

# for doing the actual multiplication 
# we use the dot method 
# for doing the actual multiplication it should follow the multiplication conditon 
# which is number of col of first matrix should be equal to the number of row of second matrix 

matrix1 = np.array([[1,2,3,4], [5,6,7,8]]) # (2,4)
matrix2 = np.array([[1,2], [3,4], [5,6], [7,8]]) # (4,2)

mul = np.dot(matrix1 , matrix2)
# the result would be of (2,2)
print(mul)

# we could also find angle .. 

```

### Vectorized Operations

```python
restaurant_types = np.array(['biryani', 'chineese', 'pizza', 'burger', 'cafe'])

# now its like we had made a method in which every string must be of capital letters
vectorized_upper = np.vectorize(str.upper)
print("Capitalized Restaurant Types: ", vectorized_upper(restaurant_types))
# Capitalized Restaurant Types:  ['BIRYANI' 'CHINEESE' 'PIZZA' 'BURGER' 'CAFE']

name = "divyansh"
Name = name.capitalize()
print(Name)
# Divyansh

Name = name.upper()
print(Name)
# DIVYANSH

# for only doing the first letter capitalize 
first_capital = np.vectorize(str.capitalize)
print("First word capitalize: ", first_capital(restaurant_types))
# First word capitalize:  ['Biryani' 'Chineese' 'Pizza' 'Burger' 'Cafe']
```
